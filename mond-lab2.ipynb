```python
import numpy as np
import matplotlib.pyplot as plt

def g(y):
    return np.exp(y) * np.sin(4 * y)

y = np.linspace(0, 4, 500)
g = g(y)

plt.plot(y,g,'r')
plt.title("Damped sine wave")
plt.show()

```

Task 2

```python
import numpy as np
import matplotlib.pyplot as plt

def g(y):
    return np.exp(y) * np.sin(4 * y)

def h(y):
    return np.exp((-3/2) * y) * np.sin(4 * y)

y = np.linspace(0, 4, 500)
g = g(y)
h = h(y)

plt.plot(y,g,'r',label = 'g(y)')
plt.plot(y,h,'black',label = 'h(y)')
plt.show()
```

Task 3

```python
import numpy as np
import matplotlib.pyplot as plt

def z(x, y):
    return np.cos(x + y) * np.sin(x*y)

x = np.linspace(-2, 2, 500)
y = np.linspace(-2, 2, 500)

X, Y = np.meshgrid(x, y)

Z = z(X, Y)

fig = plt.figure()
ax = fig.add_subplot(projection='3d')
ax.plot_surface(X, Y, Z, cmap='viridis')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')
ax.set_title('z(x,y) = cos(x+y)sin(xy)')
plt.show()
```

```python
import plotly.graph_objects as go
import numpy as np

def con(a, b):
    return a + b - np.sqrt(a**2 + b**2)

def F(x, y, z):
    return con(con(con(con(con(1 - x**2, 1 - y**2), 1 - z**2), x**2 + y**2 - 0.25), x**2 + z**2 - 0.16), y**2 + z**2 - 0.09)

X, Y, Z = np.mgrid[-1:1:0.05, -1:1:0.05, -1:1:0.05]

values = F(X,Y,Z)
print(np.amin(values))
print(np.amax(values))

fig = go.Figure(data=go.Isosurface(
    x=X.flatten(),
    y=Y.flatten(),
    z=Z.flatten(),
    value=values.flatten(),
    isomin=0,
    isomax=0,
    surface_count=1,
    caps=dict(x_show=False, y_show=False)
    ))
fig.show()
```
