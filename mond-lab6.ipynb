```python
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
from sklearn import linear_model
import matplotlib.pyplot as plt

col_id = ["fl_date","cancelled","flights","distance"]
df = pd.read_csv('/kaggle/input/mond-6-8labs/Daily-stats.xlsx',sep=';')

y = df.DISTANCE
X = df[["FLIGHTS"]]

plt.scatter(X,y)
plt.title('1')
plt.ylabel('DISTANCE')
plt.xlabel('FLIGHTS')
plt.show()

reg = linear_model.LinearRegression()

reg.fit(X,y)
y = pd.Series(np.array(reg.predict(X)).ravel())

print(reg.coef_)

plt.title('LinearRegression')
plt.scatter(X, y)
plt.show()
```

```python
reg = linear_model.Lasso(alpha=0.1)
reg.fit(X,y)

y = pd.Series(np.array(reg.predict(X)).ravel())

plt.title('Lasso')
plt.scatter(X, y)
plt.show()
```

```python
reg = linear_model.BayesianRidge()

reg.fit(X,y)

y = pd.Series(np.array(reg.predict(X)).ravel())

print(reg.coef_)

plt.title('BayesianRidge')
plt.scatter(X, y)
plt.show()
```

```python
df = pd.read_csv('/kaggle/input/mond-6-8labs/Monthly-stats.csv')

X_train = df[df.year < 2019][['year','month']]

y_train = df[df.year < 2019]['distance']

X = df[df.year == 2019][['year','month']]

plt.scatter(df['month'],df['distance'])
plt.title('1')
plt.ylabel('distance')
plt.xlabel('month')
plt.show()


reg = linear_model.LinearRegression()

reg.fit(X_train,y_train)
y = pd.Series(np.array(reg.predict(X)).ravel())

print(reg.coef_)

plt.title('Ordinary Least Squares')
plt.ylabel('distance')
plt.xlabel('month')
plt.scatter(X['month'], y)
plt.show()
```

```python
reg = linear_model.Lasso(alpha=0.1)

reg.fit(X_train,y_train)
y = pd.Series(np.array(reg.predict(X)).ravel())


plt.title('Lasso')
plt.ylabel('distance')
plt.xlabel('month')
plt.scatter(X['month'], y)
plt.show()
```

```python
reg = linear_model.BayesianRidge()

reg.fit(X_train,y_train)
y = pd.Series(np.array(reg.predict(X)).ravel())

print(reg.coef_)


plt.title('Bayesian Ridge')
plt.ylabel('distance')
plt.xlabel('month')
plt.scatter(X['month'], y)
plt.show()
```
